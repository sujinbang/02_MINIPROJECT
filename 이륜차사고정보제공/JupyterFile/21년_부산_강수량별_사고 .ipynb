```python
import pandas as pd
```

### 2021년 부산 월별 교통사고 데이터

```python
bs_ac = pd.read_excel('./minidata/21부산월별교통사고.xls')
bs_ac = bs_ac.rename(columns = bs_ac.iloc[0])
bs_ac
```

```python
bs_ac = bs_ac.loc[[1,4]]
bs_ac = bs_ac.drop(['시군구','사고년도'], axis=1)
bs_ac
```

```python
bs_ac = bs_ac.set_index(['시도'])
```

```python
bs_ac = bs_ac.drop('합계')
```

```python
bs_ac = bs_ac.drop(['합계'], axis=1)
bs_ac
```

```python
bs_ac = bs_ac.T
bs_ac
```

```python
bs_ac = bs_ac.reset_index(drop=True)
bs_ac = bs_ac.rename(columns = {'부산' : '사고건수'})
bs_ac
```

### 2021 부산 월별 강수량 데이터

```python
bs_w = pd.read_csv('./minidata/21부산월별강수량.csv', encoding='euc-kr', skiprows = 12)
bs_w = bs_w.drop(['\t지점번호','Unnamed: 8'], axis=1)
bs_w
```

```python
busan = pd.concat([bs_w, bs_ac], axis=1)
busan
```

```python
busan = busan[['일시','강수량(mm)','사고건수']]
```

```python
busan
```

```python
busan['사고건수']=busan['사고건수'].astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rc('font', family = 'Malgun Gothic', size = 10)
sns.pairplot(busan)
```

```python
plt.rc('font', family = 'Malgun Gothic', size = 10)
plt.plot(busan['강수량(mm)'],marker='o', ls='-',  
         lw=2, ms=2, mec='r',mew=2, mfc='r', label='강수량')

plt.plot(busan['사고건수'],marker='o', ls='-',  
         lw=2, ms=2, mec='b',mew=2, mfc='b', label='사고건수')
plt.title('2021년 부산 사고건수 및 강수량', fontsize=15)

plt.legend()
```

```python
import scipy.stats as stats
```

```python
# 2021년 부산 강수량 & 이륜차사고 검정
result = stats.ttest_rel(busan['강수량(mm)'], busan['사고건수'])
print('검정 결과 =', result)
print('검정 통계량 = %.3f, p-value = %.3f' %(result))
```
