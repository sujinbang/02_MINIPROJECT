```python
import pandas as pd
```

```python
data = pd.read_excel('./minidata/이륜차사고율.xlsx')
data = data.dropna()
data
```

```python
data = data.rename(columns=data.iloc[0])
data = data.drop([0], axis=0)
data = data.reset_index(drop=True)
data
```

```python
data = data.drop(['유형'], axis=1)
data = data.set_index(['구분'])
data
```

```python
data_p = data.loc[['전체사고','이륜차사고']]
data_p
```

```python
round(data_p.iloc[1][0] / data_p.iloc[0][0] * 100 , 2)
```

```python
list = []
for i in range(0,5):
    a = round(data_p.iloc[1][i] / data_p.iloc[0][i] * 100 , 2)
    list.append(a)

list
```

```python
list_df = pd.DataFrame(list)
list_df = list_df.rename(index={0:'2017년',1:'2018년',2:'2019년',3:'2020년',4:'2021년'})
```

```python
list_df = list_df.rename(columns={0:'사고율'})
```

```python
list_df
```

```python
data_p = data_p.T
```

```python
acd = pd.concat([data_p, list_df], axis=1)
acd
```

```python
acd.to_csv('./minidata/사고율.csv')
```
