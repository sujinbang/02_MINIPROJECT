```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all = pd.read_excel('./minidata/Report.xls', header=3)
```

```python
all=all[all['사고유형대분류']=='차대차']
all = all.reset_index(drop=True)
all
```

```python
all = all.loc[[0]]
all
```

```python
all_2 = all[['승용차','승합차','화물차','이륜차','자전거']]
all_2
```

```python
all_2 = all_2.astype(int)
all_3 = all_2.T
```

```python
all_3 = all_3.rename(columns={0 :'사고건수'})
```

```python
all_3 = all_3.drop(['승용차'], axis=0)
all_3
```

```python
color = ['palegreen','lightskyblue','tomato','violet']
```

```python
plt.rc("font", family = "Malgun Gothic", size=15)
plt.figure(figsize=(15,8))
plt.pie(all_3['사고건수'],labels = all_3.index, colors = color,autopct='%.1f%%',explode=(0,0,0.05,0)) #해당위치 값 돌출시키기
plt.title('2021 가해운전자 교통사고 차종', fontsize=15)
```
