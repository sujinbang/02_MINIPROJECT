# 월별 이륜차 사고건수

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_month_total = pd.DataFrame()
```

```python
df_month = pd.read_excel('./data/월별_서울_이륜차사고.xls')
df_month
```

```python
df_month = pd.read_excel('./data/월별_서울_이륜차사고.xls')
df_month = df_month.T
df_month = df_month.iloc[:,:4]
df_month = df_month.drop('시도')
df_month = df_month.drop('시군구')
df_month = df_month.drop('사고년도')
df_month = df_month.set_index(0)
df_month.columns=['사고건수','사망자수','부상자수']
df_month = df_month.drop('합계')
df_month = df_month[['사고건수']]
df_month_2018 = df_month.iloc[0:12,:]
df_month_2019 = df_month.iloc[12:24,:]
df_month_2020 = df_month.iloc[24:36,:]
df_month_2021 = df_month.iloc[36:,:]
```

```python
df_month_total = pd.concat([df_month_total,df_month_2018,df_month_2019,df_month_2020,df_month_2021],axis=1)
df_month_total
```

```python
df_month_total.columns=['2018','2019','2020','2021']
```

```python
df_month_total = df_month_total.astype(int)
df_month_total
```

```python
plt.figure(figsize=(16,10))
plt.rc("font", family = "Malgun Gothic", size=10)
sns.heatmap(df_month_total, annot=True, fmt='.0f', cmap='rocket_r')
plt.title("서울 지역 월별 이륜차 사고건수")
plt.show()
```

```python
plt.figure(figsize=(16,10))
plt.rc("font", family = "Malgun Gothic", size=15)
df_month_total.plot()
plt.title("서울 지역 월별 이륜차 사고건수")
plt.xlabel('월별')
plt.legend(df_month_total.columns)
plt.show()
```
