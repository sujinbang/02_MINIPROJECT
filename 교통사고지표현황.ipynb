```python
import pandas as pd
df = pd.DataFrame()
for i in range(0,7,1):
    file_path = f'./data/교통사고 지표 현황 ({i}).xls'
    df_file = pd.read_excel(file_path)
    df = pd.concat([df,df_file])
```

```python
df = df.reset_index(drop=True)
```

```python
df = df.dropna()
```

```python
df.columns = ['지역','유형','발생건수','?','?','?','사망자수','치사율']
```

```python
df = df[['지역', '유형', '발생건수','사망자수','치사율']]
```
